## Fehlerbehebung mit Pandas

Oft haben Daten Fehler (z.B. leere Felder, NaN, null, Prozentzeichen, Einheiten, ...)

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Data/PopulationData2.csv", decimal=".", index_col="Country").drop(columns = {"Pop/Area", "WorldPct"})
```

```python
data
```

### Fehler erkennen

isnull, notnull mit einem DataFrame um Fehlerhafte Spalten zu erkennen

```python
pd.isnull(data)
```

```python
pd.notnull(data)
```

```python
data[data["MigrantsNet"].isnull()]
```

```python
data.dropna()
```

### fillna

Füllt Fehlende Werte anhand von vorherigen/nachfolgenden Werten auf

```python
data.fillna(method="ffill")  # forwards fill (Füllt fehlende Werte vom ersten passenden Datensatz auf)
# data.fillna(method="bfill")  # backwards fill
```

ffill und bfill statt fillna

```python
data.ffill(inplace=True)
```

```python
data.bfill()
```

Fehler in der Spalte beheben anhand von Durchschnitten

```python
avg = data["MigrantsNet"].mean()
```

```python
avg
```

```python
data[data["MigrantsNet"].isnull()]["MigrantsNet"] = avg  # Wird nur "lokal" angewandt
```

Mit loc können auch Zeilen + eine Spalte ausgewählt werden

-> Für Fehlerbehandlung verwenden

```python
data.loc[data["MigrantsNet"].isnull(), "MigrantsNet"] = avg
```

```python
data
```

Drei weitere Probleme: N.A., Prozentzeichen, Datentypen (im Anschluss)
1. Suche nach N.A.
2. % entfernen in jeder Spalte
3. Datentypen anpassen

#### 1. Suche nach N.A.

```python
(data["Pop"] == "N.A.").value_counts()  # Für alle Spalten
```

```python
print("-----------")
for col in data.columns:
    print((data[col] == "N.A.").value_counts())
    print("-----------")
```

```python
for col in data.columns:
    print(f"{col}: " + str(len(data[col].groupby(data[col] == "N.A.")) > 1))
```

```python
data.loc[data["FertRate"] == "N.A.", "FertRate"] = 0
```

```python
data
```

```python
def fixNA(columnName: str, newValue):
    data.loc[data[columnName] == "N.A.", columnName] = newValue
```

```python
fixNA("MedAge", 0)
```

```python
data
```

```python
fixNA("UrbanPopPct", 0)
```

```python
data
```

#### 2. % entfernen

```python
data["Pop"].replace(" %", "", inplace=True)
```

```python
data["Pop"] = data["Pop"].replace(" %", "", regex=True)  # Hier muss regex=True angegeben werden (kein Grund)
```

```python
def fixPct(columnName: str):
    data[columnName] = data[columnName].replace(" %", "", regex=True)   # Hier muss regex=True angegeben werden (kein Grund)
```

```python
for col in data.columns:
    fixPct(col)
```

```python
data
```

##### astype(...)

Ermöglicht, eine bestimmte Spalte zu konvertieren

```python
data["YearlyChange"] = data["YearlyChange"].astype(float)  # astype auf die Spalte anwenden und dann die originale Spalte überschreiben
```

```python
data.info()
```

```python
def fixDataType(columnName: str, newType: type):
    data[columnName] = data[columnName].astype(newType)
```

```python
fixDataType("FertRate", np.float16)  # Hier können auch Numpy Datentypen verwendet werden
```

```python
fixDataType("MedAge", np.int8)
```

```python
fixDataType("UrbanPopPct", np.int8)
```

```python
fixDataType("WorldShare", np.float16)
```
