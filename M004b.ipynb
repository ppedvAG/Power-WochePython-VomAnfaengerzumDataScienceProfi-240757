### Daten aus einer Datenbank

```python
import sqlite3 as sql
```

```python
connection = sql.connect("Test.db")
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("CREATE TABLE Users (int id auto increment, name varchar(50))")
```

```python
cursor.execute("INSERT INTO Users VALUES(0, 'Max')")
```
