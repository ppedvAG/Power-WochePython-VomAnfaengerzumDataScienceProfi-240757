## Matplotlib

Daten darstellen mit verschiedenen Graphentypen

https://matplotlib.org/stable/plot_types/basic/index.html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Hier das Unterpaket pyplot importieren
```

```python
data = pd.read_csv("Data/AmazonStock.csv", thousands=",", decimal=".").drop(columns=["Date"])
```

```python
data
```

```python
data.info()
```

```python
data.plot()
```

```python
data["Open"].plot()  # X-Achse sollten Datumswerte sein
```

```python
data.set_index("Date", inplace=True)
```

```python
data["Open"].plot(figsize=(12, 4), title="AMZN Stock Data")  # Mit figsize=(B, H) die GrÃ¶ÃŸe des Plots angeben
```

### Indirekt Plotten mit Matplotlib selbst

```python
plt.plot(data["Open"])
plt.title("AMZN Stock Data")
```

```python
plt.plot(data["Open"])
plt.plot(data["Close"])
plt.plot(data["High"])
plt.plot(data["Low"])

plt.legend(["Open", "Close", "High", "Low"], loc='lower right')
plt.show()  # Einen Graphen erstellen mit Vier Datenpunkten
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(2, 2, 1)  # Zeile 1, Spalte 1, PlotNummer 1
plt.plot(data["Open"])
plt.title("Open")

plt.subplot(2, 2, 2)
plt.plot(data["Close"])
plt.title("Close")

plt.subplot(2, 2, 3)
plt.plot(data["Low"])
plt.title("Low")

plt.subplot(2, 2, 4)
plt.plot(data["High"])
plt.title("High")
```

### Globaler Modus vs. Objekt Orientierter Modus

Globaler Modus: Plotten Ã¼ber plt direkt

plt.plot, plt.title, plt.legend, ...

```python
figure, achsen = plt.subplots(2, 2)
```

```python
achsen[0, 0].plot(np.arange(100))
achsen[0, 1].plot(np.arange(50))
achsen[1, 0].plot(np.arange(75))
achsen[1, 1].plot(np.arange(25))
```

```python
figure
```

```python
achsen
```

```python
fig, ((a1_1, a1_2), (a2_1, a2_2)) = plt.subplots(2, 2)
```

```python
a1_1.plot(np.arange(100))
a1_2.plot(np.arange(50))
a2_1.plot(np.arange(75))
a2_2.plot(np.arange(25))

fig
```

### Balkendiagramm

```python
plt.bar(np.arange(10) + 0.25, np.random.random(10), width=0.5, edgecolor="black")
plt.bar(np.arange(10) - 0.25, np.random.random(10), width=0.5, edgecolor="black")
# plt.grid()
```
